### 13.4 Customizing graph

add color,linewidth,labels,title

```python
import matplotlib.pyplot as plt
import numpy as np



# TO plot a cubical  Relational graph
x = np.arange(0,5,0.1)
y = x**3
y2 = x**2

plt.plot(x,y,color = 'green',linewidth = 2, label = 'x_cube')
plt.plot(x,y2,color = 'red',linewidth = 2, label = 'x_square')


plt.ylabel("cubes and squres")
plt.xlabel("x_only")
plt.title("Test graph")


plt.legend()
plt.show()
```

Adding Grid and Axis , Also how to add random text in betn... inside the graph

```python
# To give custom axis and grid

plt.plot(x,y,color = 'green',linewidth = 2, label = 'x_cube')
plt.plot(x,y2,color = 'red',linewidth = 2, label = 'x_square')


plt.ylabel("cubes and squres")
plt.xlabel("x_only")
plt.title("Test graph")

# for custom axis
# give 4 values start and ent pt of x,start and end pt of y
plt.axis([0,10,0,200])


# to add grid
plt.grid()


#Random test in betn
plt.text(2.5,75,'Cube_line',fontsize = 12,color = 'brown')

plt.legend()
plt.show()
```

```python

import matplotlib.pyplot as plt
pop = [200,270,300,490,500,680]
sal = [12,13,14,15,16,17]
yr = [2013,2014,2015,2016,2017,2018]

co = ['green','blue','purple','yellow','cyan','grey']
plt.scatter(yr,sal,s = pop,c = co,alpha = 0.5)

plt.title('Bubble Chart')
```

TO simply add color,change shape of cricle(marker) ,

```python
import matplotlib.pyplot as plt
import numpy as np
pop = [200,270,300,490,500,680]
sal = [12,13,14,15,16,17]
yr = [2013,2014,2015,2016,2017,2018]


colo = np.arange(len(yr))

plt.scatter(yr,sal,s = pop,c = colo,alpha = 0.5,marker = '^')

plt.title('Bubble Chart')
```

 To add text beside bubbles , Showing of population of each 

```python

for i in range(len(yr)):
    plt.text(yr[i] + 0.1,sal[i] + 0.1,pop[i])


plt.scatter(yr,sal,s = pop,c = colo,alpha = 0.5,marker = 'p')

plt.title('Bubble Chart')
```

### 13.8 Pie Chart

```python
import matplotlib.pyplot as plt

c = [5,9,10,2,7,16]

plt.pie(c)
```

NOW , to add labels,texts,percentages,change color and all that

```python

c = [5,9,10,2,7,16]
labs = ['A','B','C','D','E','F']
cols = ['cyan', 'magenta', 'coral', 'brown', 'turquoise', 'pink']


# autopct is used to write the percentage
# autopct = '%.2f%%' program will do the job for you


plt.pie(c,labels = labs,colors = cols,autopct = '%.2f%%')
plt.title("Pie Chart,Dil ke taj mahal me! ")
plt.show()
```

To take a Slice out of the pie chart__{Exploding}

```python

#exploding

ex = [0.1,0.2,0.4,0,0,0]
plt.pie(c,labels = labs,colors = cols,autopct = '%.1f',explode = ex)
plt.title("Pie Chart,Dil ke taj mahal me! ")
plt.show()
```

- TO make piechart clockwise _(BY Default it's anti(counter)_Clockwise) ,
- and to Change the Starting angle _(0 Deg BY Default)

```python

# To make peiecharrt go clock clockwise
# add,counterclockwise = False   _By default it's True

# startangle = [whatever you want] Deg

plt.pie(c,labels = labs,colors = cols,autopct = '%.1f',explode = ex,counterclock = False,startangle = 128)
plt.title("Pie Chart,Dil ke taj mahal me! ")
plt.show()
```
